```python
#calculate the mean Glucose and highest Glucose values of this sample 

sample_mean_glucose = sample_dataframe['Glucose'].mean()
sample_max_glucose = sample_dataframe['Glucose'].max()

print('Sample mean value of glucose is : ' , sample_mean_glucose)
print('Sample highest value of glucose is : ',  sample_max_glucose )
```

```python
#calculate the mean Glucose and highest Glucose values of actual population

population_mean_glucose = diabetes_data['Glucose'].mean()
population_max_glucose = diabetes_data['Glucose'].max()

print('Population mean value of glucose is : ' , population_mean_glucose)
print('Population highest value of glucose is : ',  population_max_glucose )
```

```python
# Question 3 

# Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the
# population and find the average mean, standard deviation and percentile for BloodPressure and
# compare this with these statistics from the population for the same variable. Again, you should
# create charts for this comparison. Report on your findings
```

```python
#calculating the mean, standard deviation and percentile for Blood pleasure of Population

# calculate the mean, standard deviation, and percentile for BloodPressure in the population
bloodpressue_mean = diabetes_data['BloodPressure'].mean()
bloodpressue_std = diabetes_data['BloodPressure'].std()
bloodpressue_percentile = diabetes_data['BloodPressure'].quantile([0.5,0.75,0.90])

print('BloodPressure statistics in the population data:')
print(f'Mean: {bloodpressue_mean:.2f}')
print(f'Standard deviation: {bloodpressue_std:.2f}')
print(f'50th , 75th & 95th percentiles: {bloodpressue_percentile[0.50],bloodpressue_percentile[0.75],bloodpressue_percentile[0.90]}')
```

```python
#Generating the bootstrap samples 

#create empty array's to store the bootstrap samples statistic values

bloodpressure_mean_bootstrap = np.empty(500)
bloodpressure_sd_bootstrap = np.empty(500)
bloodpressure_percentile_bootstrap = np.empty((3, 500))

n_bootstrap_samples = 500
n_obs_per_bootstrap_sample = 150

for i in range(n_bootstrap_samples):
    sample_data_i = np.random.choice(diabetes_data['BloodPressure'], size=n_obs_per_bootstrap_sample, replace=True)
    bloodpressure_mean_bootstrap[i] = np.mean(sample_data_i)
    bloodpressure_sd_bootstrap[i] = np.std(sample_data_i)
    bloodpressure_percentile_bootstrap[:, i] = np.quantile(sample_data_i, [0.5, 0.75, 0.90])


```

```python
#Comparing bootstrap means with population mean

# Set the style for the plot
sns.set_style('whitegrid')

# Create a histogram of the mean of bootstrap samples blood pressure values
sns.histplot(bloodpressure_mean_bootstrap, element='step', color='green', label='Bootstrap samples mean')



# Add vertical lines at the mean of means of bootstrap samples and population mean
plt.axvline(np.mean(bloodpressure_mean_bootstrap), color='blue', linestyle='--', label='BootStrap mean')
plt.axvline(bloodpressue_mean, color='green',linestyle='--', label='population mean')

# Add labels and title to the plot
plt.xlabel('Blood Pressure')
plt.ylabel('Count')
plt.title('Blood Pressure Distribution of bootstrap samples mean vs. Population mean ' )

# Add a legend to the plot
plt.legend(loc='upper left')

#Save the mean distribution chart 
plt.savefig('C:/Users/Moham/OneDrive/Documents/PDS/3/PDS_Assignment3/analysis_result/BootstrapMeans.png')
```

```python
#Comparing bootstrap means with population sd

# Set the style for the plot
sns.set_style('whitegrid')

# Create a histogram of the sd of bootstrap samples blood pressure values
sns.histplot(bloodpressure_sd_bootstrap, element='step', color='yellow', label='Bootstrap samples sd')



# Add vertical lines at the mean of means of bootstrap samples and population mean
plt.axvline(np.mean(bloodpressure_sd_bootstrap), color='blue', linestyle='--', label='BootStrap sd')
plt.axvline(bloodpressue_std, color='green',linestyle='--', label='population sd')

# Add labels and title to the plot
plt.xlabel('Blood Pressure')
plt.ylabel('Count')
plt.title('Blood Pressure Distribution of bootstrap samples SD vs. Population SD' )

# Add a legend to the plot
plt.legend(loc='upper left')


#Save the BMI distribution chart 
plt.savefig('C:/Users/Moham/OneDrive/Documents/PDS/3/PDS_Assignment3/analysis_result/BootstrapSD.png')
```

```python
#comparing the percentiles of the bootstrap samples and population

#50th percentile
plt.subplot(1, 3, 1)
sns.barplot(x=['Bootstrap Sample ', 'Population '], y=[np.mean(bloodpressure_percentile_bootstrap[0,]), bloodpressue_percentile[0.5]], color='blue')
plt.ylabel('50th Percentile of Blood Pressure')


#75th percentile
plt.subplot(1, 3, 2)
sns.barplot(x=['Bootstrap Sample ', 'Population '], y=[np.mean(bloodpressure_percentile_bootstrap[1,]), bloodpressue_percentile[0.75]], color='green')
plt.ylabel('75th Percentile of Blood Pressure')


#90th percentile
plt.subplot(1, 3, 3)
sns.barplot(x=['Bootstrap Sample ', 'Population'], y=[np.mean(bloodpressure_percentile_bootstrap[2,]), bloodpressue_percentile[0.9]], color='yellow')
plt.ylabel('90th Percentile of Blood Pressure')

plt.title('Comparison of 50th,75th & 90th Percentile of Blood Pressure: Bootstrap Samples vs Population')



#Save the BMI distribution chart 
plt.savefig('C:/Users/Moham/OneDrive/Documents/PDS/3/PDS_Assignment3/analysis_result/BootstrapPercentiles.png')

```
